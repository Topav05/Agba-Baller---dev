```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


# Step 1: Create a new column by replacing spaces with '+'
df['Club_Search'] = df['club_name'].str.replace(' ', '+')

# Step 2: Define a function for scraping
def get_club_image(club_search):
    url = f"https://www.thesportsdb.com/search.php?s={club_search}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the image URL
    try:
        image_url = str(soup.find_all("img")[2]).split('src="')[1][:-3]
        print(image_url)
        return image_url
    except IndexError:
        return "No image found"

# Step 3: Apply the function to each club
df['Image_URL'] = df['Club_Search'].apply(get_club_image)

print(df)
```
